```python
import pandas as pd

orders = pd.read_csv("orders.csv")
drivers = pd.read_csv("drivers.csv")

orders = orders[orders["status"] == "pending"]
orders = orders[orders["distance_km"] <= 10]

drivers = drivers[drivers["available"] == 1]
```

```python
matches = pd.merge(
    orders,
    drivers,
    on="customer_zone"
)

print(matches)
```

```python
matches = matches[
    [
        "order_id",
        "driver_id",
        "customer_zone",
        "distance_km",
        "order_value",
        "max_orders"
    ]
]

print(matches.head())
```

```python
matches["priority"] = (
    matches["order_value"] /
    matches["distance_km"]
)

print(matches.head())
```

```python
matches = matches.sort_values(
    by="priority",
    ascending=False
)

print(matches.head())
```

# Driver Data Cleaning and Preparation for Food Delivery Scheduling

## Objective

The purpose of this step is to prepare the driver dataset before building a scheduling and optimization model. Real-world datasets often contain incomplete, incorrect, or unnecessary information. Therefore, the data must be cleaned and validated before it can be used for decision-making.

In a food delivery company, drivers are one of the most important resources. Before assigning orders, we need to ensure that the driver data is accurate, complete, and suitable for scheduling.

---

## Step 1: Read the Dataset

The first step is to load the driver dataset into pandas.

This converts the CSV file into a DataFrame, which is a table-like structure that allows us to analyze, clean, and manipulate data efficiently.

Reading the dataset is always the starting point of any data analysis or optimization project.

---

## Step 2: Inspect the Dataset

Before making any changes, it is important to understand the structure of the data.

During inspection, we examine:

* The first few rows of the dataset
* The column names
* The data types of each column
* The overall structure of the table

This step helps identify potential problems and provides a better understanding of the available information.

---

## Step 3: Check for Missing Values

In real-world business data, some information may be missing.

Examples include:

* Missing driver names
* Missing availability information
* Missing delivery capacity values

Missing values can cause errors during scheduling and optimization. Therefore, they must be identified before proceeding.

---

## Step 4: Remove or Handle Missing Data

Once missing values are identified, they must be handled appropriately.

Possible approaches include:

* Removing incomplete records
* Replacing missing values with suitable defaults
* Correcting invalid entries

For scheduling applications, complete and reliable data is essential because optimization models depend on accurate information.

---

## Step 5: Keep Only Available Drivers

Not every driver is available for work on a given day.

Some drivers may be:

* Off duty
* On vacation
* Sick
* Temporarily unavailable

These drivers cannot be assigned delivery orders.

Therefore, only drivers marked as available are retained in the cleaned dataset.

This ensures that the scheduling model considers only active resources.

---

## Step 6: Keep Drivers with Delivery Capacity

Each driver has a maximum number of orders they can deliver.

For example:

* Driver A may handle 4 orders
* Driver B may handle 3 orders
* Driver C may have zero capacity

Drivers with zero capacity cannot perform deliveries and should not be included in the scheduling process.

This step removes drivers who cannot contribute to the delivery operation.

---

## Step 7: Sort and Organize the Dataset

After cleaning, the data is organized in a meaningful way.

Drivers may be sorted by:

* Delivery zone
* Driver name
* Vehicle type
* Delivery capacity

Sorting improves readability and helps identify patterns in the data.

It also simplifies later analysis and scheduling tasks.

---

## Step 8: Save the Cleaned Dataset

Once all cleaning operations have been completed, the cleaned dataset is saved as a new CSV file.

This file becomes the official input for future scheduling and optimization models.

Keeping a separate cleaned file ensures that the original raw data remains unchanged and can always be referenced if needed.

---

## Why Driver Cleaning Is Important

Scheduling decisions are only as good as the data used to make them.

If unavailable drivers or incomplete records remain in the dataset, the optimization model may generate unrealistic schedules.

By cleaning the data first, we ensure that:

* Only valid drivers are considered
* Driver information is complete
* Scheduling constraints are realistic
* Optimization results are reliable

---
